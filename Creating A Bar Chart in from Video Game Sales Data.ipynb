# Creating a bar chart
### Bar charts are great for comparing numbers between categories. In this case, we will visualize the total sales in each genre to allow for quick visual comparison between groups.

```python
# resizes the figure
plt.figure(figsize=(12,8))

# code that creates the plot
plt.bar(data['genre'], data['sales'])

# rotates the genre names on the x axis
plt.xticks(rotation=90)

# creates a title at the top of the chart
plt.title('Total Video Game Sales by Genre (1980-2020)', fontsize=20);
```

```python
# resizes the figure
plt.figure(figsize=(12,8))

# code that creates the bar chart
sns.barplot(data=data, x='genre', y='sales')

# rotates the genres on the x-axis
plt.xticks(rotation=90)

# creates a new title for the chart
plt.title('Total Video Game Sales by Genre (1980-2020)', fontsize=20)

# creates a new label for the x-axis
plt.xlabel('Genre', fontsize=14)

# creates a new label for the y-axis
plt.ylabel('Total Sales', fontsize=14);
```
